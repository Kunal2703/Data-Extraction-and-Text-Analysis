```python
def get_text(url):
    
    url1 = url
    article = Article(url1, language="en")
    
    article.download() 
    article.parse() 
    article.nlp()
    
    return article.text
```

```python
for i in range(0,len(data)):
    data['URL_ID'][i] = get_text(data['URL'][i])
```

```python
data.rename({'URL_ID':'Text'},axis=1,inplace=True)
```

```python
data.head
```

```python
def transform(text):

    review = re.sub('[^a-zA-Z0-9]', ' ',text)  # except small and capital letters and numeric remove everythong.
    review = review.lower()                    # lower it.
    review = review.split()
    
    review = [word for word in review if not word in stopwords.words('english')]   # remove stopwords.
    review = ' '.join(review)
    return review


data['Transform_Text'] = data['Text'].apply(transform)
```

```python
data['word_counts'] = data['Transform_Text'].apply(lambda x: len(str(x).split()))    
```

```python
data['average number of words per sentence'] = np.nan

for i in range(0,len(data)):
    
    data['average number of words per sentence'][i] = data['word_counts'][i]/len(nltk.sent_tokenize(data['Text'][i]))
```

**COMPLEX Word Count**

```python
def complex_word_count(x):
    
    syllable = 'aeiou'
    
    t = x.split()
    
    v = []
    
    for i in t:
        words = i.split()
        c=Counter()
        
        for word in words:
            c.update(set(word))

        n = 0
        for a in c.most_common():
            if a[0] in syllable:
                if a[1] >= 2:
                    n += 1
                
        m = 0
        p = []
        for a in c.most_common():
            if a[0] in syllable:
                p.append(a[0])
        if len(p) >= 2:
            m += 1
        
        if n >= 1 or m >= 1:
            v.append(i)
            
    return len(v) 

text1 = h['Transform_Text'][1]

complex_word_count(text1)
```

```python
data['complex_count'] = np.nan

data['complex_count'] = data['Transform_Text'].apply(lambda x: complex_word_count(x))
data.head()
```

**Analysis of Readability**

```python
data['sentence length'] = np.nan
data['Average Sentence Length'] = np.nan
data['Percentage of Complex words'] = np.nan
data['Fog Index'] = np.nan


for i in range(0,len(data)):
    
    data['sentence length'][i]  =   len(nltk.sent_tokenize(data['Text'][i]))
    data['Average Sentence Length'][i] = data['word_counts'][i]/data['sentence length'][i]
    data['Percentage of Complex words'][i] = data['complex_count'][i]/data['word_counts'][i] 
    data['Fog Index'][i] = 0.4 * (data['Average Sentence Length'][i] + data['Percentage of Complex words'][i])
```
